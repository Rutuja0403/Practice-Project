```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot for happiness & Economy(GDP)


plt.title('Plot for Happiness & Economy(GDP per Capita)')
plt.rcParams['figure.figsize'] = (20, 7)
sns.scatterplot()
sns.scatterplot(x ='Happiness Score', y ='Economy (GDP per Capita)',data=df,hue='Region', s = 100)
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('Economy (GDP per Capita)')         
```

```python
#GDP contribution by region

gdp_region = df.groupby('Region')['Economy (GDP per Capita)'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP contribution by region')
plt.ylabel('')
```

```python
#Correlation map
```

```python
df.corr()
```

```python

plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True, linewidths=0.5,linecolor="black", fmt='.2f')
       
```

**From the heatmap, it is pretty evident that all the data fields mentioned from 'GDP per capita','family'**
**'Healthy life expectancy', 'Freedom', 'Generosity', 'Trust' have a strong relation between them.**



```python
#corruption by region

corruption = df.groupby('Region')[['Trust (Government Corruption)']].mean()
corruption
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
sns.scatterplot()
sns.scatterplot(x ='Trust (Government Corruption)', y ='Region',data=corruption ,hue='Region', s = 200)
plt.legend(loc = 'lower right', fontsize = '10')
plt.xlabel('Regional Corruption Index')
plt.ylabel('')         
```

```python
top_10 = df.head(10)
bottom_10 = df.tail(10)
```

```python
# Top 10 Happiest Country
plt.title('Plot for Top 10 Happiest Country')
plt.rcParams['figure.figsize'] = (10, 5)
sns.scatterplot()
sns.scatterplot(x ='Happiness Score', y ='Country',data=top_10,hue='Region', s = 200)
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('Country')         
```

```python
# Bottom 10 Happiest Country

plt.title('Plot for Bottom 10 Happiest Country')
plt.rcParams['figure.figsize'] = (10, 5)
sns.scatterplot()
sns.scatterplot(x ='Happiness Score', y ='Country',data=bottom_10,hue='Region', s = 200)
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('Country')         

```

```python
# Freedom to make life Choices vs Happiness Score
plt.figure(figsize = (8,6))
ax = sns.jointplot( x="Freedom",y='Happiness Score',data=df,kind='reg' )
plt.suptitle('Freedom to make life Choices vs Happiness Score', fontsize = 16)
plt.tight_layout()
plt.show()
```

```python
#Corruption vs Happiness Score

plt.rcParams['figure.figsize'] = (15, 7)
sns.scatterplot()
sns.scatterplot(x ='Happiness Score', y ='Trust (Government Corruption)',data=df,hue='Region', s = 200)
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('Corruption')         

```
